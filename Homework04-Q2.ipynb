### Correlation Matrix

The purpose of the correlation matrix is to understand how the variables are related to each other in our data set.
In our dataset, there's data for every given eventID (row). Elements of a column belongs to a distinct eventID. So, the columns represent the distinct properties of an event. Some examples of the columns are: Mass of the hadronic channels, azimuthal angle, pseudorapidity, etc. 

```python
%matplotlib inline
from itertools import count
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('test.csv')
corr_matrix = df.corr()
length = corr_matrix.shape[0]
```

```python
f, ax = plt.subplots(figsize = (15, 12))
corr = corr_matrix
sns.heatmap(corr, cmap = "Blues",
            xticklabels = corr.columns.values,
            yticklabels = corr.columns.values)
```
